```python
Load2016["Hour_End"] = pd.to_datetime(Load2016["Hour_End"])
```

```python
Load2017["Hour Ending"] = pd.to_datetime(Load2017["Hour Ending"])
```

```python
Load2018["HourEnding"] = pd.to_datetime(Load2018["HourEnding"])
```

```python
Load2019["HourEnding"] = pd.to_datetime(Load2019["HourEnding"])
```

```python
Load2020["HourEnding"] = pd.to_datetime(Load2020["HourEnding"])
```

```python
Load2021["DateTime"] = pd.to_datetime(Load2021["DateTime"])
```

```python
Load2017.tail()
```

```python
Load2016['Date'] = [d.date() for d in Load2016['Hour_End']]
Load2016['Time'] = [d.time() for d in Load2016['Hour_End']]

Load2017['Date'] = [d.date() for d in Load2017['Hour Ending']]
Load2017['Time'] = [d.time() for d in Load2017['Hour Ending']]

Load2018['Date'] = [d.date() for d in Load2018['HourEnding']]
Load2018['Time'] = [d.time() for d in Load2018['HourEnding']]

Load2019['Date'] = [d.date() for d in Load2019['HourEnding']]
Load2019['Time'] = [d.time() for d in Load2019['HourEnding']]

Load2020['Date'] = [d.date() for d in Load2020['HourEnding']]
Load2020['Time'] = [d.time() for d in Load2020['HourEnding']]

Load2021['Date'] = [d.date() for d in Load2021['DateTime']]
Load2021['Time'] = [d.time() for d in Load2021['DateTime']]
```

```python
Load2021.tail()
```

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html

```python
Daily_Total2016 = Load2016.groupby(by="Date").sum()
Daily_Total2017 = Load2017.groupby(by="Date").sum()
Daily_Total2018 = Load2018.groupby(by="Date").sum()
Daily_Total2019 = Load2019.groupby(by="Date").sum()
Daily_Total2020 = Load2020.groupby(by="Date").sum()
Daily_Total2021 = Load2021.groupby(by="Date").sum()
```

```python
Daily_Total2021.tail()
```

```python
#Power_Load = pd.concat([Daily_Total2016, Daily_Total2017, Daily_Total2018, Daily_Total2019, Daily_Total2020, Daily_Total2021], ignore_index=True)
Power_Load = pd.concat([Daily_Total2016, Daily_Total2017, Daily_Total2018, Daily_Total2019, Daily_Total2020, Daily_Total2021])
```
